```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2,5)
plt.plot(x,x,label='linear',color="pink")
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x')
plt.ylabel('Y')
plt.title('simple plot')
plt.legend()
```

barplot

```python
import statistics as st
fig,ax=plt.subplots()
movie=['comedy','horror','romance','thiller','action','scifi']
rep=[5,10,12,7,8,12]
res_mean=st.mean(rep)
res_sd=st.stdev(rep)
ax.bar(movie,rep,yerr=res_sd)
ax.set_ylabel("responeses")
ax.set_xticklabels(movie,rotation=90)
fig.set_size_inches([7,5])
plt.show()
```

muptiple and stacked bar plot

```python
fig,ax=plt.subplots()
movie=['comedy','horror','romance','thiller','action','scifi']
m_rep=[5,10,12,7,8,12]
w_rep=[4,6,8,10,12,14]
ax.bar(movie,m_rep,width=0.5)
ax.bar(movie,w_rep,bottom=m_rep,width=0.5)#ax.bar(movie,w_rep)
ax.set_ylabel("responeses")
ax.set_xticklabels(movie,rotation=90)
fig.set_size_inches([7,5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-3.0,3.0,100)
y=np.linspace(-3.0,3.0,100)
x1,y1=np.meshgrid(x,y)
z=np.sqrt(x1**2+y1**2)
fig,ax=plt.subplots(1,1)
cp=ax.contourf(x1,y1,z)#countour
fig.colorbar(cp)
plt.show()
```

```python

x=np.array([5,4,3,2,1])
y=np.array([5,4,3,2,1])
plt.plot(x,y,marker="*",color="blue")
plt.annotate('mininum point',xy=(1,1),xytext=(2,1),arrowprops=dict(facecolor='black',shrink=0.05))
plt.annotate('maximum point',xy=(5,5),xytext=(3,5),arrowprops=dict(facecolor='black',shrink=0.05))
plt.show()
```

```python
x=np.array([5,4,3,2,1])
y=np.array([5,4,3,2,1])
plt.plot(x,y,marker="*",color="blue")
plt.annotate('mininum point',xy=(1,1),xytext=(2,1),arrowprops=dict(facecolor='black',shrink=0.05,connectionstyle="arc3,rad=0.3"))
plt.annotate('maximum point',xy=(5,5),xytext=(3,5),arrowprops=dict(facecolor='black',shrink=0.05,connectionstyle="arc3,rad=0.3"))
plt.show()
```
